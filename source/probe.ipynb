```python
import pandas as pd
```

```python
data = pd.read_csv("../dataset/dataset.csv")

data.describe()
```

```python
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Extract all plagiarized and non-plagiarized samples
plagiarized = data[data["Plagiarized"] == 1]
non_plagiarized = data[data["Plagiarized"] == 0]

# Randomly sample 700 non-plagiarized
non_plagiarized_sample = non_plagiarized.sample(n=700, random_state=42)

# Randomly sample 700 plagiarized (or all if less than 700)
plagiarized_sample = plagiarized.sample(n=min(700, len(plagiarized)), random_state=42)

# Concatenate and shuffle
balanced_data = pd.concat([non_plagiarized_sample, plagiarized_sample]).sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
import re

def remove_imports(code):
    # Remove all 'import ...;' statements (with optional spaces/tabs before)
    return re.sub(r'\s*import[^;]*;', '', code)

# Apply to both File1 and File2 columns
balanced_data_no_imports = balanced_data.copy()
balanced_data_no_imports['File1'] = balanced_data_no_imports['File1'].apply(remove_imports)
balanced_data_no_imports['File2'] = balanced_data_no_imports['File2'].apply(remove_imports)
# Save the balanced dataset without imports
balanced_data_no_imports.to_csv("../dataset/dataset_v2.csv", index=False)
```
